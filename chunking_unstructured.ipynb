```python
from unstructured.partition.text import partition_text
```

### React Typescript

```python
with open("examples/page.tsx", "r") as f:
    text = f.read()
    tsx_elements = partition_text(text=text)
    tsx_elements_basic = partition_text(
        text=text, chunking_strategy="basic", max_characters=1000
    )
    tsx_elements_by_title = partition_text(
        text=text, chunking_strategy="by_title", max_characters=1000
    )

print("Basic: ", [len(el.text) for el in tsx_elements_basic])
print("By title: ", [len(el.text) for el in tsx_elements_by_title])
```

Look into what code are being categorize as 'Title'

```python
limit = 0
for el in tsx_elements:
    if limit > 10:
        break
    if el.category == "Title":
        print(el.text)
        print("==")
        limit += 1
```
